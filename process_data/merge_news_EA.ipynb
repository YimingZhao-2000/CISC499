```python
# import the libraries you need, make sure they are installed within your Python env
import pandas as pd

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"


pd.options.display.width = None
```

```python
ea_label = pd.read_csv('ea_labeled.csv')
ea_date = pd.read_csv('../raw_data/ea_engadget.csv')
```

```python
ea_date
ea_label
```

```python
ea_complete = ea_label
ea_complete['Date'] = ea_date["date"]
```

```python
# use the to_datetime() method to convert the Date column to a datetime format
ea_complete['Date'] = pd.to_datetime(ea_complete['Date'], format='%m.%d.%Y')
ea_complete['score'] = pd.to_numeric(ea_complete['score'], errors='coerce')
```

```python
# check which rows have NaN values in column 'A'
print(ea_complete[ea_complete['score'].isna()])
```

```python
ea_complete.dtypes
```

```python
ea_complete.drop(columns=['title'], inplace=True)
# group rows by date and calculate mean value of each group
ea_complete = ea_complete.groupby('Date').mean()
```

```python
mother_data = pd.read_csv('filled_EA.csv')
mother_data['Date'] = pd.to_datetime(mother_data['Date'])
```

```python
mother_data = mother_data.merge(ea_complete, left_on='Date', right_on='Date', how='outer')
mother_data.sort_values(by=['Date'], inplace=True)
```

```python
mother_data = mother_data[mother_data['Date'] <= "2019-12-30"]
```

```python
mother_data[mother_data['score'].notna() & mother_data['Close_Price'].isna()]
```

```python
for idx, row in mother_data.iterrows():
    # check if the Close_Price value is NaN and the score value is not NaN
    if pd.isna(row['Close_Price']) and not pd.isna(row['score']):
        # check if the score value in the next row is not NaN and if it is not the last row
        if idx < len(mother_data) - 1 and not pd.isna(mother_data.loc[idx+1, 'score']):
            mother_data.loc[idx+1, 'score'] = mother_data.loc[idx+1, 'score'] / 2 + row['score'] / 2
        else:
            mother_data.drop(idx, inplace=True)
```

```python
mother_data[mother_data['score'].isna() & mother_data['Close_Price'].notna()]
```

```python
col, row = mother_data.shape
```

```python
mother_data['Days_After_Last_News'] = [0]*col
```

```python
# Because the first row of column ''score'' is NaN, thus for writing code, we just assume the emotion score as 0 - 0 means no effect on emotion
mother_data['score'][0] = 0
```

```python
mother_data
```

```python
for idx, row in mother_data.iterrows():

    if pd.isna(row['score']):
        mother_data.loc[idx, 'score'] = mother_data.loc[idx-1, 'score']
        mother_data.loc[idx, 'Days_After_Last_News'] = mother_data.loc[idx-1, 'Days_After_Last_News']+1

mother_data
```

```python
mother_data.to_csv('merged_news_EA.csv', index=False)
```
