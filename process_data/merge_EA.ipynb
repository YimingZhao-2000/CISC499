**Note**: This code is under processing_raw_data folder, and the data that we are merging are located under raw data folder

```python
# read all the csv that we are merging
ea_ychart = pd.read_csv("../raw_data/ea_price_data.csv")
ndx_100 = pd.read_csv("../raw_data/NDX100.csv")
spx_data = pd.read_csv("../raw_data/SPX_data.csv")
tbond_10 = pd.read_csv("../raw_data/I10YTCMR_data.csv") # data include year of 2020
ea_pe_ratio = pd.read_csv("../raw_data/ea_pe_ratio_data.csv") # data include year of 2020
```

```python
time_stamp = pd.date_range(start='1/1/2010', end='12/31/2019')
```

```python
# Check ychart data types
ea_ychart.dtypes
ea_ychart.head(10)
```

```python
ea_ychart["Date"] =  pd.to_datetime(ea_ychart["Date"])
```

```python
ea_ychart.dtypes
ea_ychart.head(10)
```

```python
# concate timestamp with y-chart
data_joint = pd.DataFrame({"DATE": time_stamp})
data_joint = data_joint.merge(ea_ychart, left_on="DATE", right_on="Date", how="left")
data_joint = data_joint.drop("Date", axis=1)
```

```python
data_joint.dtypes
```

```python
ndx_100["Period"] = pd.to_datetime(ndx_100["Period"])
data_joint = data_joint.merge(ndx_100, left_on="DATE", right_on="Period", how="left")
data_joint = data_joint.drop("Period", axis=1)
```

```python
data_joint
```

```python
spx_data.head(10)
```

```python
spx_data["Period"] = pd.to_datetime(spx_data["Period"])
data_joint = data_joint.merge(spx_data, left_on="DATE", right_on="Period", how="left")
data_joint = data_joint.drop("Period", axis=1)
```

```python
data_joint
```

```python
tbond_10.head(10)
```

```python
tbond_10["Period"] = pd.to_datetime(tbond_10["Period"])
data_joint = data_joint.merge(tbond_10, left_on="DATE", right_on="Period", how="left")
data_joint = data_joint.drop("Period", axis=1)
```

```python
data_joint.head(10)
```

```python
ea_pe_ratio.head(10)
ea_pe_ratio["Period"] = pd.to_datetime(ea_pe_ratio["Period"])
data_joint = data_joint.merge(ea_pe_ratio, left_on="DATE", right_on="Period", how="left")
data_joint = data_joint.drop("Period", axis=1)
data_joint.head(100)
```
