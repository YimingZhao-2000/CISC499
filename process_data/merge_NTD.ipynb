```python
# import the libraries you need, make sure they are installed within your Python env
import pandas as pd
import seaborn as sns
import numpy as np
import plotly.express as px
import missingno as msno
from scipy.stats import skew, norm
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from yellowbrick.regressor import ResidualsPlot
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"


pd.options.display.width = None
print("Libraries imported succesfully")
```

**Note**: This code is under processing_raw_data folder, and the data that we are merging are located under raw data folder

```python
# read all the csv that we are merging
ntd_ychart = pd.read_csv("../raw_data/ntd_price_data.csv")
ndx_100 = pd.read_csv("../raw_data/NDX100.csv")
spx_data = pd.read_csv("../raw_data/SPX_data.csv")
tbond_10 = pd.read_csv("../raw_data/I10YTCMR_data.csv")
ntd_pe_ratio = pd.read_csv("../raw_data/ntd_pe_ratio_data.csv") 
```

```python
time_stamp = pd.date_range(start='1/1/2010', end='12/31/2019')
```

```python
data_joint = pd.DataFrame({"DATE": time_stamp})
ntd_ychart["Date"] =  pd.to_datetime(ntd_ychart["Date"])
data_joint = data_joint.merge(ntd_ychart, left_on="DATE", right_on="Date", how="left")
data_joint = data_joint.drop("Date", axis=1)
ndx_100["Period"] = pd.to_datetime(ndx_100["Period"])
data_joint = data_joint.merge(ndx_100, left_on="DATE", right_on="Period", how="left")
data_joint = data_joint.drop("Period", axis=1)
spx_data["Period"] = pd.to_datetime(spx_data["Period"])
data_joint = data_joint.merge(spx_data, left_on="DATE", right_on="Period", how="left")
data_joint = data_joint.drop("Period", axis=1)
tbond_10["Period"] = pd.to_datetime(tbond_10["Period"])
data_joint = data_joint.merge(tbond_10, left_on="DATE", right_on="Period", how="left")
data_joint = data_joint.drop("Period", axis=1)
ntd_pe_ratio["Period"] = pd.to_datetime(ntd_pe_ratio["Period"])
data_joint = data_joint.merge(ntd_pe_ratio, left_on="DATE", right_on="Period", how="left")
data_joint = data_joint.drop("Period", axis=1)
```
